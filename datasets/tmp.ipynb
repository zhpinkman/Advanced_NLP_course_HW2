```python
import pandas as pd
import numpy as np
```

```python
with open("4dim.train.txt") as f:
    data = f.read().splitlines()
```

```python
all_lengths = []
for sent in data:
    all_lengths.append(len(sent.split()))
```

```python
np.mean(all_lengths), np.max(all_lengths), np.median(all_lengths)
```

```python
from matplotlib.pyplot import hist


hist(all_lengths)
```
