```python
import pandas as pd
import numpy as np
from matplotlib.pyplot import hist
import matplotlib.pyplot as plt
```

```python
with open("odia.train.txt") as f:
    data = f.read().splitlines()
all_lengths = []
for sent in data:
    all_lengths.append(len(sent.split()))
print(np.mean(all_lengths), np.median(all_lengths))
hist(all_lengths)
plt.show()
```
